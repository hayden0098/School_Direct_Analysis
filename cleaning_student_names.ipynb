```python
#load and read file
#import pandas function
import pandas as pd
#load
student_file_load = "Resources/students_complete.csv"

#read
student_data_df = pd.read_csv(student_file_load)
```

```python
name_list = student_data_df["student_name"].tolist()
name_list
```

```python
for name in name_list:
    print(name.split(), len(name.split()))
len(name_list)
```

```python
#store pre/suffix name to a list
pre_suf_fix=[]
for name in name_list:
    if len(name.split())>=3:
        pre_suf_fix.append(name)
len(pre_suf_fix)
print(pre_suf_fix)
```

```python
prefix=[]
for name in pre_suf_fix:
    if len(name.split()[0]) <=4:
        prefix.append(name.split()[0])

suffix=[]
for name in pre_suf_fix:
    if len(name.split()[-1])<=3:
        suffix.append(name.split()[-1])
```

```python
#get unique item in list
prefix_name = set(prefix)
print(prefix_name)

suffix_name = set(suffix)
print(suffix_name)
```

```python
#replace method to replace prefixe with empty string of string
#make list of common pre/suffixes
prefixes_suffixes = ["Dr. ", "Mr. ","Ms. ", "Mrs. ", "Miss ", " MD", " DDS", " DVM", " PhD", " IV", " II", " III", " V", " Jr."]

for name in prefixes_suffixes:
    student_data_df["student_name"] = student_data_df["student_name"].str.replace(name,"")
student_data_df
```

```python
# Put the cleaned students' names in another list.
student_names = student_data_df["student_name"].tolist()
student_names
```

```python
#check the length for student name
# Create a new list and use it for the for loop to iterate through the list.
students_fixed = []

# Use an if statement to check the length of the name.

# If the name is greater than or equal to 3, add the name to the list.

for name in student_names:
    if len(name.split()) >= 3:
        students_fixed.append(name)

# Get the length of the students' names that are greater than or equal to 3.
len(students_fixed)
```
